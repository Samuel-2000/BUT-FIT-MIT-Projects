## Pomocné funkce
Zde jsou definované pomocné vykreslovací funkce. Jedná se tedy hlavně o vykreslování skupiny daných grafu pro všechny atributy.

```python
def generate_boxplots(cmp_attribute):
  for attribute in df.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=attribute, y=cmp_attribute, data=df)
    plt.title(f'Boxplot pro {attribute} v závislosti na {cmp_attribute}')
    plt.xticks(rotation=45)
    plt.show()
    plt.figure(figsize=(6, 3))
    sns.violinplot(x=attribute, y=cmp_attribute, data=df)
    plt.title(f'Houslovy pro {attribute} v závislosti na {cmp_attribute}')
    plt.xticks(rotation=45)
    plt.show()

def course_boxplots(cmp_attribute):
  for attribute in df.select_dtypes(include=['object']).drop(columns=['Course']).columns:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x='Course', y=cmp_attribute, data=df, hue=attribute)
    plt.title(f'Boxplot pro {attribute} v závislosti na {cmp_attribute} podle oboru')
    plt.xticks(rotation=45)
    plt.show()

def generate_coutplot(cmp_attribute):
  for attribute in df.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(6, 3))
    sns.countplot(x=attribute, hue=cmp_attribute, data=df)
    plt.title(f'Countplot pro {attribute} v závislosti na {cmp_attribute}')
    plt.xticks(rotation=45)
    plt.show()

def generate_catplot(cmp_attribute):
  for attribute in df.select_dtypes(include=['object']).drop(columns=['Course']).columns:
    plt.figure(figsize=(6, 3))
    sns.catplot(x='Course', y=cmp_attribute, hue=attribute, data=df, kind='bar')
    plt.title(f'Barplot pro {attribute} v závislosti na {cmp_attribute} podle oboru')
    plt.xticks(rotation=45)
    plt.show()
```

# Detekce odlehlých hodnot
Podle již předchnozích grafů jde vidět, že dává smysl kontrolovat pro odlehlé hodnoty atributy věk, gpa a počet kreditů v semestru.

## Age
Prvně použijeme mezodu zscore pro vyhledání odlehlých hodnot pro atribut age a ty následně zobrazíme. Zároveň zobrazíme boxplot pro atribut age kde je také možné vidět odlehlé hodnoty.

```python
from scipy.stats import zscore

z_scores_age = zscore(df['Age'])
outliers_age = (z_scores_age > 3) | (z_scores_age < -3)
outliers_data_age = df[outliers_age]
outliers_data_age
```

```python
sns.boxplot(x=df['Age'])
plt.title('Boxplot pro Age')
plt.show()
```

## CGPA
Stejný postup následně uděláme aji pro atribut cgpa, kde tedy prvně vyhledáme odlehlé hodnoty pomocí metodz zscore a následně ještě zobrazíme boxplot kde jdou odlehlé hodnoty vidět.

```python
no_nan_cgpa = df.dropna(subset=['CGPA'])
z_scores_cgpa = zscore(no_nan_cgpa['CGPA'])
outliers_cgpa = (z_scores_cgpa > 3) | (z_scores_cgpa < -3)
outliers_data_cgpa = no_nan_cgpa[outliers_cgpa]
outliers_data_cgpa
```

```python
sns.boxplot(x=df['CGPA'])
plt.title('Boxplot pro CGPA')
plt.show()
```

## Doplnění chybějící hodnoty
Pro atribut cgpa nechybí velké množství dat a jedná se o numerický atribut. Proto dává smysl chybějícíc hodnoty doplnit a tedy nahradit ho za průměrnou hodnotu cgpa, kde použijeme knihovnu sklearn. Následně zobrazíme řádky kde některé měli chybějící hodnotu pro cgpa.

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
df['CGPA'] = imputer.fit_transform(df[['CGPA']])

df.isna().sum().sort_values(ascending=False)
```

# Odstranění odlehlých hodnot
Pro odstranění odlehlých hodnot použijeme již vyhledané odlehlé data ze sekce Detekce odlehlých hodnot. Odstraníme tedy odlehlé hodnoty pro věk a cgpa a následně zobrazíme začátek dat kde byli některé tyto odlehlé hodnoty.

```python
z_scores_age = zscore(df['Age'])
outliers_age = (z_scores_age > 3) | (z_scores_age < -3)

df = df[~outliers_age]

z_scores_cgpa = zscore(df['CGPA'])
outliers_cgpa = (z_scores_cgpa > 3) | (z_scores_cgpa < -3)

df = df[~outliers_cgpa]

df[0:50]
```

# Diskretizace numerických atributů
Pro diskretizaci numerických atributů použijeme knihovnu sklearn, kde zvolíme počet rozdělení intervalu na 5 a požijeme strategii quantile, které rozdělí data tak, že v každém intervalu je stejný počet hodnot.

```python
from sklearn.preprocessing import KBinsDiscretizer

numeric_features = df.select_dtypes(include=[np.number]).columns
kbins_discretizer = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='quantile')
df1 = df.copy()
df1[numeric_features] = kbins_discretizer.fit_transform(df[numeric_features])
df1
```

# Transformace kategorických atributů na numerické
Pro transformaci kategorických atributů na numerické použijeme metodu get_dummies která pro všechny hodnoty kategorických atributů vytvoří nové atributy a tu hodnotu který daný řádek nabýval se označí 1 a ostatní 0.

```python
df2 = pd.get_dummies(df, dtype=int)
df2
```

# Normalizace atributů
Pro normalizaci použijeme knihovnu sklearn. Kde pracujeme s datama které mají všechny atributy převedeny na numerické. Aplikujeme na ně normalizaci MinMax která všechny hodnoty v atributu na inverval 0 až 1.

```python
from sklearn.preprocessing import MinMaxScaler

numeric_features = df2.select_dtypes(include=[np.number]).columns
scaler = MinMaxScaler()
df2_norm = df2.copy()
df2_norm[numeric_features] = scaler.fit_transform(df2[numeric_features])

df2_norm
```
